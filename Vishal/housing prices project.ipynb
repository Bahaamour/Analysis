**Differences in Unemployment Rate Across Time Between Education Clusters**

We can see that the two clusters follow the same general trend across the years from 2008-2018.

For instance both graphs see a rise in unemployment rates that culminates around 2009-2011 before steadily declining for the rest of the period.

Also note how large the variation between states is even within the educated and uneducated states within this period graphically. We can formalize this variation with variance calculations.


```python
display(df_ue.var(axis=1))
display(df_e.var(axis=1))
```

**Assessing Correlation between the Clusters**

```python
var_ue=df_ue.var(axis=1)
var_ue_df=pd.DataFrame(var_ue,columns=["Uneducated States Variance"])

var_e=df_e.var(axis=1)
var_e_df=pd.DataFrame(var_e,columns=["Educated States Variance"])

var_ue_df.hvplot(xlabel="years",ylabel="Variance in Unemployment Rate",label="Uneducated States Variance")*var_e_df.hvplot(xlabel="years",ylabel="Variance in Unemployment Rate",label="Educated States Variance")
```

Examining the differences empirically, the empirical variances confirm the graphical variation per each year even within clusters

For instance, within the uneducated cluster, 2009 saw a variance value of 2.057 which is more than 5x the value in 2018. 

Also note that while the variance within the educated cluster fluctuates up and down somewhat, the variance within the uneducated cluster is more pronounced directionally (there is a clear rise from 2008-2009) before dipping down and somewhat plateuing (even though there are some minor fluctuations). This plateau from 2008-2009 and subsequent steady decline does seem to suggest that the uneducated cluster was more sensitive to financial crisis 2008-2009.

```python
var_ue_df["Uneducated States Variance"].corr(var_e_df["Educated States Variance"])
```

The correlation value of -0.413 is negative and close to halfway between 0 and -1, suggesting that there is a moderate (not strong) inversely proportionate relationship in variance values between the two clusters. This (-0.413) correlation is weak relative to what would be seen if the correlation value were close to -1.0 or 1.0, and suggests that there are more uncorrelated segments of the graph than correlated ones, as a balanced picture would be accompanied by a correlation value of 0.5 or -0.5. However, the modicum of inverse correlation must be acknowledged, as exemplified in the intervals of 2010-2011, 2012-2013, 2013-2014, etc where while one graph increases, the other one decreases (often not at the same rate).

```python
var_ue_df["Uneducated States Variance"].cov(var_e_df["Educated States Variance"])
```

This covariance value of -0.07361113636363635 is very close to 0, which can be interpreted as there being an absence of a linear relationship between the two clusters' variance. 

A possible interpretation, given the -0.413 correlation value and the -0.07316 covariance is that although there might be a probablistic distribution in which the variance values of the clusters belong to, it's dubuous and uncertain. Thus, I'm not inclined to say with certainty whether the two clusters are correlated in variance.

```python
df_ue_mean=pd.DataFrame(df_ue.mean(axis=1),columns=['mean unemployment (uneducated states)'])
display(df_ue_mean)
display(df_ue_mean.var())
```

```python
df_e_mean=pd.DataFrame(df_e.mean(axis=1),columns=['mean unemployment (educated states)'])
display(df_e_mean)
```

```python
df_e_mean.hvplot(xlabel="years",ylabel="mean unemployment rate (%)",label='educated states')*df_ue_mean.hvplot(xlabel="years",ylabel="mean unemployment rate (%)",label='uneducated states')
```

```python
df_e_mean['mean unemployment (educated states)'].corr(df_ue_mean['mean unemployment (uneducated states)'])
```

This extremely high correlation of .993 indicates that the two clusters were equally sensitive to the events affecting employment rates across time. Moreover, it is genuinely surprising it similar unemployment rates are across clusters, especially since the variance correlation had the opposite sign and was far smaller in absolute magnitude. This disparity might suggest that collapsing the 5 unemployment values of the items (states) in a cluster into one metric (mean) might result in informational loss and oversimplify the picture

Yet, this does suggest that there still is an underlying temporal probablistic distribution under which both clusters' mean unemployment rate values fall into.
